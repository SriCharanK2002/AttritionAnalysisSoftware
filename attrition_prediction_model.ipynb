Handling Categorical Values

```python
df = data.copy()
```

```python
Attrition_le = LabelEncoder()
BusinessTravel_le = LabelEncoder()
Department_le = LabelEncoder()
EducationField_le = LabelEncoder()
Gender_le = LabelEncoder()
JobRole_le = LabelEncoder()
MaritalStatus_le = LabelEncoder()
Over18_le = LabelEncoder()
OverTime_le = LabelEncoder()
```

```python
df['Attrition'] = Attrition_le.fit_transform(df['Attrition'])
df['BusinessTravel'] = BusinessTravel_le.fit_transform(df['BusinessTravel'])
df['Department'] = Department_le.fit_transform(df['Department'])
df['EducationField'] = EducationField_le.fit_transform(df['EducationField'])
df['Gender'] = Gender_le.fit_transform(df['Gender'])
df['JobRole'] = JobRole_le.fit_transform(df['JobRole'])
df['MaritalStatus'] = MaritalStatus_le.fit_transform(df['MaritalStatus'])
df['Over18'] = Over18_le.fit_transform(df['Over18'])
df['OverTime'] = OverTime_le.fit_transform(df['OverTime'])
```

Designing the model

```python
x = mdf.drop(['Attrition'] , axis = 1).values
y = mdf['Attrition' ].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size= 0.25 , random_state=42)
[x_train.shape, y_train.shape]
```

```python
ro_scaler = RobustScaler()
x_train = ro_scaler.fit_transform(x_train)
x_test = ro_scaler.fit_transform(x_test)
[x_train.shape, x_test.shape]
```

```python
model_scores = {}
```

Logistic Regression Model

```python
lr = LogisticRegression(C = 100)
lr.fit(x_train , y_train)
score = lr.score(x_test , y_test)
model_scores['LogisticRegression'] = score
```

Decision Tree Classifier

```python
dtc = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5)
dtc.fit(x_train, y_train)
score = dtc.score(x_test, y_test)
model_scores["DecisionTreeClassifier"] = score
```

K Nearest Neighbours

```python

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
score = knn.score(x_test, y_test)
model_scores["KNeighborsClassifier"] = score
```

Naive Bayes

```python

nb = GaussianNB()
nb.fit(x_train, y_train)
score = nb.score(x_test, y_test)
model_scores["NaiveBayes"] = score
```

Stochastic Gradient Descent

```python

sgd = SGDClassifier(loss="modified_huber", shuffle=True, random_state=100)
sgd.fit(x_train, y_train)
score = sgd.score(x_test, y_test)
model_scores["StochasticGradientDescent"] = score
```

Random Forest Classifier

```python
rfc = RandomForestClassifier(n_estimators=200, oob_score=True, n_jobs=5, random_state=101, max_features=None, min_samples_leaf=20)
rfc.fit(x_train, y_train)
score = rfc.score(x_test, y_test)
model_scores["RandomForestClassifier"] = score
```

Support Vector Machines

```python
svm = SVC(kernel='linear', C=0.1, random_state=101)
svm.fit(x_train, y_train)
score = svm.score(x_test, y_test)
model_scores["SupportVectorMachines"] = score
```

```python
model_scores
```

```python
max_model = max(model_scores, key= lambda x: model_scores[x])
print("Best Model for Attrition Prediction: {a} \nWith a score of {b}".format(a = max_model,b = model_scores[max_model]))
```

#### Let's preprocess the data a little

```python
pipe_model_scores = dict()
```

Logistic Regression

```python
pipe_lr = Pipeline(steps = [('scaler',StandardScaler()),('lr',LogisticRegression(C = 100))])
pipe_lr.fit(x_train , y_train)
score = pipe_lr.score(x_test , y_test)
pipe_model_scores['LogisticRegression'] = score
```

Decision Tree Classifier

```python
pipe_dtc = Pipeline([('scaler',StandardScaler()),('dtc',DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5))])
pipe_dtc.fit(x_train, y_train)
score = pipe_dtc.score(x_test, y_test)
pipe_model_scores["DecisionTreeClassifier"] = score
```

KNN Classifier

```python
pipe_knn = Pipeline([('scaler',StandardScaler()),('knn',KNeighborsClassifier(n_neighbors=3))]) 
pipe_knn.fit(x_train, y_train)
score = pipe_knn.score(x_test, y_test)
pipe_model_scores["KNeighborsClassifier"] = score
```

Naive Bayes Classifier

```python
pipe_nb = Pipeline([('scaler',StandardScaler()),('nb',GaussianNB())])
pipe_nb.fit(x_train, y_train)
score = pipe_nb.score(x_test, y_test)
pipe_model_scores["NaiveBayes"] = score
```

Stochastic Gradient Descent

```python
pipe_sgd = Pipeline([('scaler',StandardScaler()),('sgd',SGDClassifier(loss="modified_huber", shuffle=True, random_state=100))])
pipe_sgd.fit(x_train, y_train)
score = pipe_sgd.score(x_test, y_test)
pipe_model_scores["StochasticGradientDescent"] = score
```

Random Forest Classifier

```python
pipe_rfc = Pipeline([('scaler',StandardScaler()),('rfc',RandomForestClassifier(n_estimators=200, oob_score=True, n_jobs=5, random_state=101, max_features=None, min_samples_leaf=20))])
pipe_rfc.fit(x_train, y_train)
score = pipe_rfc.score(x_test, y_test)
pipe_model_scores["RandomForestClassifier"] = score
```

Support Vector Machines

```python
pipe_svm = Pipeline([('scaler',StandardScaler()),('svm',SVC(kernel='linear', C=0.1, random_state=101))])
pipe_svm.fit(x_train, y_train)
score = pipe_svm.score(x_test, y_test)
pipe_model_scores["SupportVectorMachines"] = score
```

```python
pipe_model_scores
```

```python
max_model = max(model_scores, key= lambda x: model_scores[x])
print("Best Model for Attrition Prediction even after scaling is {a}\nWith a score of {b}".format(a = max_model,b = model_scores[max_model]))
```
